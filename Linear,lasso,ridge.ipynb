```python
numeric_cols = train.select_dtypes(include=['float64', 'int64']).columns
#copy data
train_cleaned = train.copy()

# in numeric_cols only clean of outliet
for col in numeric_cols:
    Q1 = train_cleaned[col].quantile(0.25)
    Q3 = train_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    train_cleaned = train_cleaned[(train_cleaned[col] >= lower_bound) & (train_cleaned[col] <= upper_bound)]

# final result after remove outlier
train = train_cleaned
```

```python

numeric_features = train_cleaned.select_dtypes(include=['int64', 'float64'])

corr_matrix = numeric_features.corr()
saleprice_corr = corr_matrix['SalePrice'].sort_values(ascending=False)
strong_corr = saleprice_corr[abs(saleprice_corr) > 0.3]

plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix.loc[strong_corr.index, strong_corr.index],
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)
plt.title(" Correlated Features with SalePrice", fontsize=14)
plt.show()
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
```

```python
X = train.drop(['SalePrice'], axis=1)
y = train['SalePrice']
```

```python
X_categorical = X.select_dtypes(include=['object'])# convret from obj to encoded true or false
X_categorical_encoded = pd.get_dummies(X_categorical, drop_first=True)
```

```python
X_categorical_encoded.head()
```

```python
X_numeric = X[numeric_cols.drop('SalePrice', errors='ignore')]
X_final = pd.concat([X_numeric , X_categorical_encoded], axis=1) # after encoded merge feature
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
 # Scaling the features to have mean = 0 and standard deviation = 1
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge(alpha=0.0)  #0.0 by defult( linear reg)
ridge.fit(X_train_scaled, y_train)
```

```python
coff=ridge.coef_
coff
```

```python
y_pred = ridge.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error


# MAE is the average of the absolute differences between predicted and actual values
print("MAE:", mean_absolute_error(y_test, y_pred))
# RMSE is the square root of MSE, in the same units as the target
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
# difference between actual and predicted values (lower is better)
print("MSE:", mean_squared_error(y_test, y_pred))
# indicates how well the model explains the variance in the target (closer to 1 is better)
print("R²:", r2_score(y_test, y_pred))
```

```python
# plot Actual vs Predicted SalePrice
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, color='skyblue', edgecolor='black')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
plt.xlabel('Actual SalePrice')
plt.ylabel('Predicted SalePrice')
plt.title('Actual vs Predicted SalePrice')
plt.grid(True)
plt.show() # fig show is overfit solve problem by use ridge
```
